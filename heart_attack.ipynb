```python
import pandas as pd
```

```python
heart_attack_pred = pd.read_csv("heart_attack_prediction_dataset.csv").drop(columns=["Patient ID", "Country", "Continent", "Hemisphere"])
heart_attack_pred.columns = map(str.lower, heart_attack_pred)
heart_attack_pred.head()
```

```python
heart_attack_pred['bmi'] = round(heart_attack_pred['bmi'], 1)
heart_attack_pred.head()
```

```python
medical_insurance = pd.read_csv("medical_insurance.csv")
medical_insurance['bmi'] = round(medical_insurance['bmi'], 1)
medical_insurance.head()
```

```python
heart_attack_pred['sex'] = heart_attack_pred['sex'].str.lower()
df = heart_attack_pred.merge(medical_insurance, on=['age', 'sex', 'bmi'])
df.head()
```

```python
df[['systolic blood pressure', 'diastolic blood pressure']] = df['blood pressure'].str.split('/', expand=True)
df = df.drop(columns=['blood pressure'])
df.head()
```
